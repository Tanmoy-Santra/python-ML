<b><i>Take x values and plot the cos(x) and tan(x) in a graph using matplotlib.</i></b>
<hr>
<br>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-1,1,0.001) 
y=np.cos(x)
z=np.tan(x)
```

```python
plt.grid() 
plt.plot(x,y,color="blue")
plt.plot(x,z,color="green")
plt.show()
```
