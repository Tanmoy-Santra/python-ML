<b><i>Take a data set in a data frame and plot the bar chart example may be year of Production and the Production Quantity. (Take help of Pandas)</i></b><br>
<hr>
<br>

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = {
    'Year' : [2000,2005,2010,2015,2020],
    'Qnt' : [10,8,15,17,20]
}
```

```python
df = pd.DataFrame(dataset)
x = df['Year']
y = df['Qnt']
```

```python
plt.bar(x,y,color="red")
plt.title("Year of Production vs Production Quantity Bar Chart")
plt.xlabel('Year of Production')
plt.ylabel('Production Quantity')
plt.show()
```
