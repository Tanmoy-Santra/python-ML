```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('D:\CSE_2021_22\employee.csv')
x = df['Name']
y = df['Salary']
```

```python
plt.bar(x,y)
plt.xlabel("Employee Name")
plt.ylabel("Employee Salary")
plt.title("Employee Name vs Employee Salary Bar Chart")
```
