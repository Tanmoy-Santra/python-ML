<h2><b><u>Multiple box plot</u> :</b></h2>
<p><b>A multiple box plot, also known as a grouped box plot or side-by-side box plot, is a graphical representation used to compare and display the distribution of a continuous variable for different categories or groups. It is an extension of the traditional box plot, which is used to show the summary statistics of a single dataset, such as the median, quartiles, and potential outliers. In a multiple box plot, you can compare the distribution of the same variable across different categories, making it easier to identify differences, similarities, and outliers 
between the groups</b>.</p>

```python
import matplotlib.pyplot as plt
import pandas as pd
data = { 
'Parameter1': [10, 15, 20, 25, 30], 
'Parameter2': [5, 12, 18, 22, 28], 
'Parameter3': [8, 14, 21, 27, 33], 
} 
df = pd.DataFrame(data) 
df.to_csv('multiple_boxplot.csv', index=False) 
df = pd.read_csv('multiple_boxplot.csv')
plt.figure(figsize=(6,4)) 
df.boxplot(column=['Parameter1', 'Parameter2', 'Parameter3'], grid=False) 
plt.title('Multiple Box Plots for Different Parameters') 
plt.ylabel('Values') 
plt.show()
```
