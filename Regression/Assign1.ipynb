```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv("D:\CSE_2021_22\Position_Salaries.csv")
x=df[['Level']].values
y=df['Salary'].values
```

```python
lin_reg=LinearRegression().fit(x,y)
y_pred=lin_reg.predict(x)
plt.scatter(x,y,color="blue")
plt.plot(x,y_pred,"r-")
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
print("R2 score for linear regression",metrics.r2_score(y,y_pred))
```

<br><br><br>

```python
poly_reg=PolynomialFeatures(degree=5)
xpoly=poly_reg.fit_transform(x)
lin_reg=LinearRegression().fit(xpoly,y)
y_pred=lin_reg.predict(xpoly)
plt.scatter(x,y,color="blue")
plt.plot(x,y_pred,"r-")
plt.title('Polinomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
print("R2 score for polynomial regression",metrics.r2_score(y,y_pred))
```
